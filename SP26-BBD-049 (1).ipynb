```python
# Product inventory dictionary
inventory = {}

# Function to add a new product
def add_product(name, quantity, price):
    if name in inventory:
        print(f"{name} already exists. Use update function instead.")
    else:
        inventory[name] = {
            "quantity": quantity,
            "price": price
        }
        print(f"{name} added successfully.")

# Function to update product quantity
def update_quantity(name, quantity):
    if name in inventory:
        inventory[name]["quantity"] = quantity
        print(f"{name} quantity updated to {quantity}.")
    else:
        print(f"{name} not found in inventory.")

# Function to calculate total inventory value
def total_inventory_value():
    total = 0
    for product in inventory.values():
        total += product["quantity"] * product["price"]
    return total

# Function to generate low-stock report
def low_stock_report(threshold=5):
    low_stock_items = {}
    for name, details in inventory.items():
        if details["quantity"] < threshold:
            low_stock_items[name] = details
    return low_stock_items
add_product("Laptop", 10, 800)
add_product("Mouse", 3, 20)
add_product("Keyboard", 2, 50)
update_quantity("Mouse", 6)
print("\nTotal Inventory Value:", total_inventory_value())

print("\nLow Stock Items:")
low_items = low_stock_report(threshold=5)
for name, details in low_items.items():
    print(f"{name} -> Quantity: {details['quantity']}, Price: {details['price']}")
```

```python
# Dictionary to store student information
students = {
    "Ali": {"Math": 85, "English": 78, "Science": 92},
    "Sara": {"Math": 90, "English": 88, "Science": 95},
    "John": {"Math": 70, "English": 75, "Science": 80}
}

# Function to calculate average grade for each student
def calculate_averages(students):
    averages = {}
    for name, subjects in students.items():
        avg = sum(subjects.values()) / len(subjects)
        averages[name] = avg
    return averages

# Function to find highest and lowest grades in a subject
def subject_extremes(students, subject):
    grades = []
    
    for name, subjects in students.items():
        if subject in subjects:
            grades.append((name, subjects[subject]))
    
    if not grades:
        return "Subject not found."
    
    highest = max(grades, key=lambda x: x[1])
    lowest = min(grades, key=lambda x: x[1])
    
    return highest, lowest

# Function to sort students by average grades
def sort_by_average(students):
    averages = calculate_averages(students)
    sorted_students = sorted(averages.items(), key=lambda x: x[1], reverse=True)
    return sorted_students

# Calculate averages
averages = calculate_averages(students)
print("Average Grades:")
for name, avg in averages.items():
    print(f"{name}: {avg:.2f}")

# Find highest and lowest in Math
highest, lowest = subject_extremes(students, "Math")
print("\nMath Highest:", highest)
print("Math Lowest:", lowest)

# Sort students by average
sorted_students = sort_by_average(students)
print("\nStudents Sorted by Average:")
for name, avg in sorted_students:
    print(f"{name}: {avg:.2f}")
```
